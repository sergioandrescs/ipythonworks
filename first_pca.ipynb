```python
%pylab inline
cova=[]
count=-1
barxy=[]
data= loadtxt('first_pca.dat')
for i in range(shape(data)[1]):
    barxy.append((mean(data[:,i])))
scatter(data[:,0],data[:,1])

for k in range(len(barxy)**2):
    u=k%len(barxy)
    a=0
    if(u==0):
        cova.append([])
        count+=1
    for i in range(len(data[:,0])):
        a+=(data[i,(count)]-barxy[count])*(data[i,u]-barxy[u])
    cova[count].append(a/(len(data[:,0])-1))
eigen=linalg.eig(cova)[1]
print eigen[0]
print cova
m0=eigen[0][0]/eigen[0][1]
xlines=linspace(0.5,3,1000)
plot(xlines,xlines*m0)
m1=eigen[1][0]/eigen[1][1]
plot(xlines,4.2+xlines*m1)

```
